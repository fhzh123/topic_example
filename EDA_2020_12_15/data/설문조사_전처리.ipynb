```python
eng_sec1 = ['timestamp', 'sex', 'age', 'job', 'background', 'agency_know', 'agency_view']
eng_sec2 = ['past_views'] # '미디어 앤 아트' 기획 전시 사전 경험
eng_sec3 = ['frequency', 'porpose', 'companion', 'motivation', 'info_path'] # Part 1. 전시 관람 경험
eng_sec4 = ['chap1_favor', 'chap1_reason', 'chap2_favor', 'chap2_reason', 'chap3_favor', 'chap31_reason'] # Part 2. 전시물 선호도
eng_sec5 = ['tot_satisfy', 'sat_point'] # 전시 만족도
eng_sec6 = ['sat_storytell', 'sat_mapping', 'sat_oled', 'sat_emotion', 'sat_conven', 'sat_safety'] # 항목 별 만족도 수준
eng_sec7 = ['fat_move', 'fat_dense', 'fat_tech'] # 전시방법 및 기술 피로도 수준
eng_sec8 = ['emotion', 'immersion'] # 전시의 몰입적 특성
eng_sec9 = ['complement', 'revisit', 'recommend', 'covid_after', 'reserve', 'covid_effect', 'covid_freq', 'share'] # 향후 관람 의사
eng_sec10 = ['share_method'] # 관람 경험 공유 방법에 대한 문항
eng_sec11 = ['text']
```

```python
eng_name = eng_sec1+eng_sec2+eng_sec3+eng_sec4+eng_sec5+eng_sec6+eng_sec7+eng_sec8+eng_sec9+eng_sec10+eng_sec11
```

```python
len(eng_name)
```

```python
columns2eng = {}
eng2columns = {}
for c, e in zip(data.columns, eng_name):
    columns2eng[c] = e
    eng2columns[e] = c
```

```python
data.rename(columns = columns2eng, inplace = True)
```

```python
data.head()
```

```python
data.info()
```

```python
data_labelencode = data.copy()
classes_name = {}
cate_name = []
cont_name = []
#encoder = LabelEncoder()
for e in eng_name[1:7]+eng_name[8:-1]:
    if is_string_dtype(data[e]): # object dtype 이면... label encoding 하기
        cate_name.append(e)
        encoded, classes = pd.factorize(data[e])
        data_labelencode[e] = encoded
        classes_name[e] = classes
        #data_labelencode[e] = encoder.fit_transform(data[e].values)
        #value_name[e] = encoder.classes_
    else:
        cont_name.append(e)
```

```python
data_labelencode.head()
```

```python
data_labelencode.to_csv('./survey_answer_encoded.csv', index=False)
```
